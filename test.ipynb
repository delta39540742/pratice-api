```python

import requests
import base64
import json
#TODO: change <your pinggy link> into real link (ex: "https://odyrb-34-16-208-161.run.pinggy-free.link")
API_URL = "<your pinggy link>"

### test get basic info
response_basic_info = requests.get(API_URL)
# data_basic_info = response_basic_info.json()
print(f"basic info: {response_basic_info.json()}")

### test get health
response_health_info = requests.get(f"{API_URL}/health")
print(f"health: {response_health_info.json()}")

### test generate image
payload = {
    "prompt": "The dog is singing."
}

# Gửi request
response = requests.post(f"{API_URL}/generate", params=payload, timeout=120)
# Kiểm tra mã trạng thái HTTP (200 là OK)
if response.status_code == 200:
    data = response.json()
    img_b64 = data["image_base64"]
    
    # KIỂM TRA: Nếu chuỗi có tiền tố "data:image/...", cần cắt bỏ nó
    if "," in img_b64:
        img_b64 = img_b64.split(",")[1]
    
    # Giải mã và lưu file
    img_binary = base64.b64decode(img_b64)
    with open("result_test.png", "wb") as f:
        f.write(img_binary)
        
    print("Đã lưu ảnh thành công: result_test.png")
else:
    # Xử lý khi lỗi (404, 500, v.v.)
    print(f"Lỗi HTTP: {response.status_code}")
    try:
        data = response.json()
        print(f"Thông báo từ server: {data.get('message', 'Không có chi tiết')}")
    except:
        print("Không thể đọc nội dung lỗi.")
```
